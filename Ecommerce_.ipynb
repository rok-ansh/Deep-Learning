```python
# To remove the extra space after the sentence - strip()
df['Description']=df['Description'].str.strip()
df.dropna(axis = 0, subset = 'InvoiceNo', inplace = True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
# This is to remove any extra character in invoice eg C580480
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
# Lets create a basket based on Invoice customer buyed a product

basket = (df[df['Country'] == 'France'].groupby(['InvoiceNo','Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)).set_index('InvoiceNo')
```
